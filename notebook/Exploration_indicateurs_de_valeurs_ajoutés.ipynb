### Analyse des performances : Top/Flop par indicateur

### Heatmap de performance avec normalisation

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.impute import SimpleImputer
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Analyser les valeurs manquantes d'abord
def analyser_na(df, colonnes):
    """Analyse le pourcentage de NA par colonne"""
    na_analysis = {}
    for col in colonnes:
        na_count = df[col].isna().sum()
        na_percent = (na_count / len(df)) * 100
        na_analysis[col] = {
            'na_count': na_count,
            'na_percent': na_percent,
            'valeurs_manquantes': df[df[col].isna()].index.tolist()
        }
    return na_analysis

# Sélectionner les indicateurs clés
perf_cols = ['taux_de_reussite', 'taux_de_mentions', 'taux_d_acces_2nde_bac',
            'part_de_mentions_tb_avec_félicitations', 'part_de_mentions_tb_sans_félicitations',
            'part_de_mentions_b', 'part_de_mentions_ab']

df_perf_bac = df_region[perf_cols].copy()

# 🔍 ANALYSE DES VALEURS MANQUANTES
na_analysis = analyser_na(df_perf_bac, perf_cols)

print("📊 ANALYSE DES VALEURS MANQUANTES")
print("="*50)
for col, info in na_analysis.items():
    print(f"\n{col}:")
    print(f"  Valeurs manquantes: {info['na_count']}/{len(df_perf_bac)} ({info['na_percent']:.1f}%)")
    if info['valeurs_manquantes']:
        print(f"  Régions concernées: {', '.join(info['valeurs_manquantes'])}")

# 🎯 STRATÉGIE D'IMPUTATION
def imputation_intelligente(df, seuil_na=30):
    """
    Impute les valeurs manquantes selon une stratégie adaptée
    """
    df_imputed = df.copy()
    colonnes_imputees = []
    colonnes_supprimees = []

    for col in df.columns:
        na_percent = (df[col].isna().sum() / len(df)) * 100

        if na_percent == 0:
            # Pas de NA, rien à faire
            continue
        elif na_percent <= seuil_na:
            # Imputation avec SimpleImputer
            print(f"✅ Imputation de '{col}' ({na_percent:.1f}% NA)")

            # Essayer différentes stratégies selon la distribution
            if df[col].std() / df[col].mean() < 0.2:  # Faible variabilité -> médiane
                strategy = 'median'
            else:  # Forte variabilité -> moyenne
                strategy = 'mean'

            imputer = SimpleImputer(strategy=strategy)
            df_imputed[col] = imputer.fit_transform(df[[col]]).flatten()
            colonnes_imputees.append(f"{col} (stratégie: {strategy})")

        else:
            # Trop de NA, on supprime ou on traite spécialement
            print(f"⚠️  '{col}' a {na_percent:.1f}% NA (> {seuil_na}%) - Suppression recommandée")
            colonnes_supprimees.append(col)

    return df_imputed, colonnes_imputees, colonnes_supprimees

# Appliquer l'imputation
df_imputed, cols_imputees, cols_supprimees = imputation_intelligente(df_perf_bac, seuil_na=30)

print(f"\n🔧 RÉSULTAT DE L'IMPUTATION")
print("="*35)
print(f"Colonnes imputées: {len(cols_imputees)}")
for col in cols_imputees:
    print(f"  • {col}")

if cols_supprimees:
    print(f"\nColonnes à supprimer (trop de NA): {len(cols_supprimees)}")
    for col in cols_supprimees:
        print(f"  • {col}")

    # Supprimer les colonnes avec trop de NA
    df_final_clean = df_imputed.drop(columns=cols_supprimees)
else:
    df_final_clean = df_imputed

print(f"\nDataframe final: {df_final_clean.shape}")
```

```python
# 🚀 VERSION AVANCÉE AVEC CHOIX DE STRATÉGIES
def imputation_avancee(df, seuil_na=30):
    """
    Imputation avancée avec choix automatique de la meilleure stratégie
    """
    from sklearn.impute import KNNImputer
    from sklearn.experimental import enable_iterative_imputer
    from sklearn.impute import IterativeImputer

    df_results = {}
    strategies_info = {}

    for col in df.columns:
        na_percent = (df[col].isna().sum() / len(df)) * 100

        if na_percent == 0:
            df_results[col] = df[col].copy()
            strategies_info[col] = "Pas de NA"
            continue

        if na_percent > seuil_na:
            print(f"⚠️ {col}: {na_percent:.1f}% NA (ignoré)")
            continue

        # Tester différentes stratégies
        strategies = {}

        # 1. SimpleImputer avec différentes méthodes
        for method in ['mean', 'median']:
            imputer = SimpleImputer(strategy=method)
            strategies[f'Simple_{method}'] = imputer.fit_transform(df[[col]]).flatten()

        # 2. KNN Imputer (si pas trop de colonnes)
        if df.shape[1] <= 10:
            try:
                knn_imputer = KNNImputer(n_neighbors=3)
                # Utiliser toutes les colonnes numériques pour KNN
                df_numeric = df.select_dtypes(include=[np.number])
                strategies['KNN'] = knn_imputer.fit_transform(df_numeric)[:, df_numeric.columns.get_loc(col)]
            except:
                pass

        # Choisir la meilleure stratégie (celle qui préserve le mieux la distribution)
        original_std = df[col].std()
        best_strategy = None
        best_score = float('inf')

        for strategy_name, imputed_values in strategies.items():
            # Score basé sur la préservation de l'écart-type
            new_std = pd.Series(imputed_values).std()
            score = abs(original_std - new_std)

            if score < best_score:
                best_score = score
                best_strategy = strategy_name
                df_results[col] = imputed_values

        strategies_info[col] = f"{best_strategy} (NA: {na_percent:.1f}%)"
        print(f"✅ {col}: {best_strategy} choisie ({na_percent:.1f}% NA)")

    return pd.DataFrame(df_results, index=df.index), strategies_info

# Appliquer l'imputation avancée
df_advanced_imputed, strategies_used = imputation_avancee(df_perf_bac)

print(f"\n🎯 STRATÉGIES UTILISÉES")
print("="*30)
for col, strategy in strategies_used.items():
    print(f"{col:<40} : {strategy}")
```

```python
from sklearn.preprocessing import MinMaxScaler, RobustScaler, StandardScaler

# Analysons d'abord les distributions de tes données
def analyser_distributions(df):
    """Analyse les distributions pour choisir le bon scaler"""
    stats = {}

    for col in df.columns:
        data = df[col].dropna()

        # Statistiques descriptives
        Q1 = data.quantile(0.25)
        Q3 = data.quantile(0.75)
        IQR = Q3 - Q1

        # Détection des outliers (méthode IQR)
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        outliers = data[(data < lower_bound) | (data > upper_bound)]

        # Coefficient de variation
        cv = data.std() / data.mean() if data.mean() != 0 else 0

        # Test de normalité (Shapiro-Wilk pour petits échantillons)
        from scipy import stats as scipy_stats
        if len(data) <= 50:
            _, p_value = scipy_stats.shapiro(data)
            is_normal = p_value > 0.05
        else:
            is_normal = None

        stats[col] = {
            'mean': data.mean(),
            'median': data.median(),
            'std': data.std(),
            'min': data.min(),
            'max': data.max(),
            'Q1': Q1,
            'Q3': Q3,
            'IQR': IQR,
            'outliers_count': len(outliers),
            'outliers_percent': (len(outliers) / len(data)) * 100,
            'cv': cv,
            'skewness': data.skew(),
            'is_normal': is_normal,
            'outliers_values': outliers.tolist()
        }

    return stats

# Analyser les données du bac
stats_bac = analyser_distributions(df_perf_bac.dropna())

print("📊 ANALYSE DES DISTRIBUTIONS")
print("="*50)
for col, stat in stats_bac.items():
    print(f"\n{col}:")
    print(f"  Outliers: {stat['outliers_count']} ({stat['outliers_percent']:.1f}%)")
    print(f"  Coefficient variation: {stat['cv']:.3f}")
    print(f"  Skewness: {stat['skewness']:.3f}")
    if stat['outliers_count'] > 0:
        print(f"  Valeurs aberrantes: {[f'{x:.1f}' for x in stat['outliers_values'][:3]]}")
```

```python
# Comparaison des 3 scalers principaux
def comparer_scalers(df):
    """Compare l'effet des différents scalers"""

    scalers = {
        'MinMaxScaler': MinMaxScaler(),
        'RobustScaler': RobustScaler(),
        'StandardScaler': StandardScaler()
    }

    # Préparer les données (sans NaN pour cette démo)
    df_clean = df.dropna()

    fig, axes = plt.subplots(3, len(df_clean.columns), figsize=(20, 12))

    for i, (scaler_name, scaler) in enumerate(scalers.items()):
        # Normaliser les données
        df_scaled = pd.DataFrame(
            scaler.fit_transform(df_clean),
            index=df_clean.index,
            columns=df_clean.columns
        )

        for j, col in enumerate(df_clean.columns):
            ax = axes[i, j]

            # Histogramme
            ax.hist(df_scaled[col], bins=10, alpha=0.7, color=f'C{i}')
            ax.set_title(f'{scaler_name}\n{col}', fontsize=10)
            ax.set_ylabel('Fréquence')

            # Statistiques
            ax.text(0.02, 0.98,
                   f'Min: {df_scaled[col].min():.2f}\n' +
                   f'Max: {df_scaled[col].max():.2f}\n' +
                   f'Médiane: {df_scaled[col].median():.2f}',
                   transform=ax.transAxes,
                   verticalalignment='top',
                   bbox=dict(boxstyle='round', facecolor='white', alpha=0.8),
                   fontsize=8)

    plt.tight_layout()
    plt.suptitle('Comparaison des scalers sur les données du baccalauréat',
                 fontsize=16, fontweight='bold', y=1.02)
    plt.show()

# Appliquer la comparaison
comparer_scalers(df_perf_bac[['taux_de_reussite', 'taux_de_mentions', 'taux_d_acces_2nde_bac']])
```
