```python
plt.figure(figsize=(12, 8))

# Calculer les médianes pour diviser en quadrants
median_ratio = df_nb['ratio'].median()
median_ips = df_nb['ips'].median()

# Scatter plot avec couleurs par quadrant
colors = []
categories = []
for _, row in df_nb.iterrows():
    if row['ratio'] >= median_ratio and row['ips'] >= median_ips:
        colors.append('#e74c3c')  # Rouge: Forte densité + IPS élevé
        categories.append('Forte densité - IPS élevé')
    elif row['ratio'] >= median_ratio and row['ips'] < median_ips:
        colors.append('#f39c12')  # Orange: Forte densité + IPS faible
        categories.append('Forte densité - IPS faible')
    elif row['ratio'] < median_ratio and row['ips'] >= median_ips:
        colors.append('#27ae60')  # Vert: Faible densité + IPS élevé
        categories.append('Faible densité - IPS élevé')
    else:
        colors.append('#3498db')  # Bleu: Faible densité + IPS faible
        categories.append('Faible densité - IPS faible')

plt.scatter(df_nb['ratio'], df_nb['ips'],
           c=colors, s=100, alpha=0.8, edgecolors='black', linewidth=0.5)

# Lignes de référence
plt.axhline(y=median_ips, color='gray', linestyle='--', alpha=0.5)
plt.axvline(x=median_ratio, color='gray', linestyle='--', alpha=0.5)

# Annotations
for i, row in df_nb.reset_index().iterrows():
    plt.annotate(row['libelle_region'],
                (row['ratio'], row['ips']),
                xytext=(5, 5), textcoords='offset points',
                fontsize=9)

# Légende personnalisée
from matplotlib.patches import Patch
legend_elements = [
    Patch(facecolor='#e74c3c', label='Forte densité - IPS élevé'),
    Patch(facecolor='#f39c12', label='Forte densité - IPS faible'),
    Patch(facecolor='#27ae60', label='Faible densité - IPS élevé'),
    Patch(facecolor='#3498db', label='Faible densité - IPS faible')
]
plt.legend(handles=legend_elements, loc='upper right')

plt.title('Analyse par quadrants: IPS vs Densité d\'élèves',
          fontsize=16, fontweight='bold')
plt.xlabel('Ratio élèves/établissement', fontsize=12)
plt.ylabel('IPS moyen', fontsize=12)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd

# Créer les catégories (version corrigée)
df_analysis = df_nb.reset_index().copy()
median_ratio = df_nb['ratio'].median()
median_ips = df_nb['ips'].median()

# Méthode plus simple et robuste
def categoriser_region(row):
    if row['ratio'] >= median_ratio and row['ips'] >= median_ips:
        return 'Forte densité - IPS élevé'
    elif row['ratio'] >= median_ratio and row['ips'] < median_ips:
        return 'Forte densité - IPS faible'
    elif row['ratio'] < median_ratio and row['ips'] >= median_ips:
        return 'Faible densité - IPS élevé'
    else:
        return 'Faible densité - IPS faible'

df_analysis['categorie'] = df_analysis.apply(categoriser_region, axis=1)

# Afficher par catégorie
categories = [
    'Forte densité - IPS élevé',
    'Forte densité - IPS faible',
    'Faible densité - IPS élevé',
    'Faible densité - IPS faible'
]

for cat in categories:
    print(f"\n{cat.upper()}:")
    regions = df_analysis[df_analysis['categorie'] == cat]['libelle_region'].tolist()
    print(f"Régions: {', '.join(regions)}")
```

```python
niveaux_data = eff_colleges_mayotte[['region_academique','rentree_scolaire','6èmes_total','5èmes_total','4èmes_total','3èmes_total']].groupby(['region_academique','rentree_scolaire']).sum()
niveaux_data

# Reset index pour faciliter la manipulation
niveaux_reset = niveaux_data.reset_index()

# Créer le suivi de cohorte
def suivre_cohorte(data, annee_debut_6eme):
    """Suit une cohorte depuis la 6ème"""
    cohorte = {}

    # 6ème à l'année de début
    sixieme = data[(data['rentree_scolaire'] == annee_debut_6eme)]
    if not sixieme.empty:
        cohorte['6ème'] = {'annee': annee_debut_6eme, 'effectifs': sixieme['6èmes_total'].sum()}

    # 5ème l'année suivante
    cinquieme = data[(data['rentree_scolaire'] == annee_debut_6eme + 1)]
    if not cinquieme.empty:
        cohorte['5ème'] = {'annee': annee_debut_6eme + 1, 'effectifs': cinquieme['5èmes_total'].sum()}

    # 4ème deux ans après
    quatrieme = data[(data['rentree_scolaire'] == annee_debut_6eme + 2)]
    if not quatrieme.empty:
        cohorte['4ème'] = {'annee': annee_debut_6eme + 2, 'effectifs': quatrieme['4èmes_total'].sum()}

    # 3ème trois ans après
    troisieme = data[(data['rentree_scolaire'] == annee_debut_6eme + 3)]
    if not troisieme.empty:
        cohorte['3ème'] = {'annee': annee_debut_6eme + 3, 'effectifs': troisieme['3èmes_total'].sum()}

    return cohorte

# Identifier les cohortes possibles
annees_disponibles = sorted(niveaux_reset['rentree_scolaire'].unique())
cohortes_possibles = [annee for annee in annees_disponibles if annee + 3 in annees_disponibles]

plt.figure(figsize=(14, 10))

# Graphique principal : évolution des cohortes
for region in niveaux_reset['region_academique'].unique():
    data_region = niveaux_reset[niveaux_reset['region_academique'] == region]

    for i, annee_debut in enumerate(cohortes_possibles):
        cohorte = suivre_cohorte(data_region, annee_debut)

        if len(cohorte) >= 2:  # Au moins 2 niveaux disponibles
            niveaux_ordre = ['6ème', '5ème', '4ème', '3ème']
            effectifs = [cohorte[niveau]['effectifs'] for niveau in niveaux_ordre if niveau in cohorte]
            positions = list(range(len(effectifs)))

            # Couleur différente par cohorte et style par région
            couleur = plt.cm.Set3(i / len(cohortes_possibles))
            style = '-' if region == 'Mayotte' else '--'

            plt.plot(positions, effectifs,
                    color=couleur, linestyle=style,
                    marker='o', linewidth=3, markersize=8,
                    label=f'{region} - Cohorte {annee_debut}',
                    alpha=0.8)

plt.title('Évolution des cohortes d\'élèves par niveau\n(Suivi longitudinal)',
          fontsize=16, fontweight='bold')
plt.xlabel('Niveau scolaire', fontsize=12)
plt.ylabel('Nombre d\'élèves', fontsize=12)
plt.xticks(range(4), ['6ème', '5ème', '4ème', '3ème'])
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```
