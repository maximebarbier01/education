```python
df_nb = merge_ips_eff.groupby('libelle_region').agg({
    'nombre_d_eleves':'sum'
    ,'uai':'count'
    ,'ips':'mean'
}).rename(columns={'uai': 'nombre_d_etablissements'}).round(2)
df_nb['ratio'] = df_nb['nombre_d_eleves'] / df_nb['nombre_d_etablissements']
df_nb.sort_values('ratio',ascending=False)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Le dataframe est d√©j√† tri√©, parfait !
plt.figure(figsize=(12, 10))
bars = plt.barh(df_nb.index, df_nb['ratio'],
                color='#e67e22', alpha=0.8, edgecolor='white', linewidth=0.8)

# Ajouter les valeurs √† droite des barres
for i, bar in enumerate(bars):
    width = bar.get_width()
    plt.text(width + 2, bar.get_y() + bar.get_height()/2,
             f'{width:.0f}', ha='left', va='center', fontweight='bold')

plt.title('Nombre moyen d\'√©l√®ves par √©tablissement par r√©gion\n(Classement d√©croissant)',
          fontsize=16, fontweight='bold', pad=20)
plt.xlabel('√âl√®ves par √©tablissement', fontsize=12)
plt.ylabel('R√©gion', fontsize=12)
plt.grid(axis='x', alpha=0.3)

# Ligne de r√©f√©rence (moyenne nationale)
moyenne_nationale = df_nb['ratio'].mean()
plt.axvline(x=moyenne_nationale, color='red', linestyle='--', linewidth=2,
            label=f'Moyenne nationale: {moyenne_nationale:.0f}')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

plt.figure(figsize=(12, 8))

# Scatter plot avec r√©gression
sns.scatterplot(data=df_nb.reset_index(),
                x='ratio', y='ips',
                s=100, alpha=0.7, color='#3498db')

# Ajouter les noms des r√©gions
for i, row in df_nb.reset_index().iterrows():
    plt.annotate(row['libelle_region'],
                (row['ratio'], row['ips']),
                xytext=(5, 5), textcoords='offset points',
                fontsize=9, alpha=0.8)

# Ligne de r√©gression
z = np.polyfit(df_nb['ratio'], df_nb['ips'], 1)
p = np.poly1d(z)
plt.plot(df_nb['ratio'], p(df_nb['ratio']),
         color='red', linestyle='--', linewidth=2, alpha=0.8)

# Calcul et affichage de la corr√©lation
correlation = df_nb['ratio'].corr(df_nb['ips'])
plt.text(0.05, 0.95, f'Corr√©lation: {correlation:.3f}',
         transform=plt.gca().transAxes,
         bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.8),
         fontsize=12, fontweight='bold')

plt.title('Relation entre IPS et densit√© d\'√©l√®ves par √©tablissement',
          fontsize=16, fontweight='bold')
plt.xlabel('Ratio √©l√®ves/√©tablissement', fontsize=12)
plt.ylabel('IPS moyen', fontsize=12)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(14, 8))

# Scatter avec taille proportionnelle au nombre d'√©l√®ves
scatter = plt.scatter(df_nb['ratio'], df_nb['ips'],
                     s=df_nb['nombre_d_eleves']/1000,  # Taille des bulles
                     c=df_nb['ips'],
                     cmap='RdYlBu',
                     alpha=0.7,
                     edgecolors='black',
                     linewidth=0.5)

# Colorbar
cbar = plt.colorbar(scatter)
cbar.set_label('IPS moyen', fontsize=12)

# Annotations pour les r√©gions
for i, row in df_nb.reset_index().iterrows():
    plt.annotate(row['libelle_region'],
                (row['ratio'], row['ips']),
                xytext=(5, 5), textcoords='offset points',
                fontsize=8, alpha=0.9)

plt.title('IPS vs Ratio √©l√®ves/√©tablissement\n(Taille des bulles = nombre d\'√©l√®ves)',
          fontsize=16, fontweight='bold')
plt.xlabel('Ratio √©l√®ves/√©tablissement', fontsize=12)
plt.ylabel('IPS moyen', fontsize=12)
plt.grid(True, alpha=0.3)

# L√©gende pour la taille des bulles
sizes = [50000, 200000, 500000]
labels = ['50k', '200k', '500k √©l√®ves']
legend_elements = [plt.scatter([], [], s=s/1000, c='gray', alpha=0.6,
                              edgecolors='black', linewidth=0.5)
                  for s in sizes]
plt.legend(legend_elements, labels,
          title="Nombre d'√©l√®ves", loc='upper right')

plt.tight_layout()
plt.show()
```

### Mayotte au college

```python
eff_colleges_mayotte = eff_colleges[eff_colleges['region_academique'] == 'MAYOTTE']
eff_colleges_mayotte
```

```python
niveaux_data = eff_colleges_mayotte[['region_academique','rentree_scolaire','6√®mes_total','5√®mes_total','4√®mes_total','3√®mes_total']].groupby(['region_academique','rentree_scolaire']).sum()
niveaux_data

# Reset index pour faciliter la manipulation
niveaux_reset = niveaux_data.reset_index()

# Cr√©er le suivi de cohorte
def suivre_cohorte(data, annee_debut_6eme):
    """Suit une cohorte depuis la 6√®me"""
    cohorte = {}

    # 6√®me √† l'ann√©e de d√©but
    sixieme = data[(data['rentree_scolaire'] == annee_debut_6eme)]
    if not sixieme.empty:
        cohorte['6√®me'] = {'annee': annee_debut_6eme, 'effectifs': sixieme['6√®mes_total'].sum()}

    # 5√®me l'ann√©e suivante
    cinquieme = data[(data['rentree_scolaire'] == annee_debut_6eme + 1)]
    if not cinquieme.empty:
        cohorte['5√®me'] = {'annee': annee_debut_6eme + 1, 'effectifs': cinquieme['5√®mes_total'].sum()}

    # 4√®me deux ans apr√®s
    quatrieme = data[(data['rentree_scolaire'] == annee_debut_6eme + 2)]
    if not quatrieme.empty:
        cohorte['4√®me'] = {'annee': annee_debut_6eme + 2, 'effectifs': quatrieme['4√®mes_total'].sum()}

    # 3√®me trois ans apr√®s
    troisieme = data[(data['rentree_scolaire'] == annee_debut_6eme + 3)]
    if not troisieme.empty:
        cohorte['3√®me'] = {'annee': annee_debut_6eme + 3, 'effectifs': troisieme['3√®mes_total'].sum()}

    return cohorte

# Identifier les cohortes possibles
annees_disponibles = sorted(niveaux_reset['rentree_scolaire'].unique())
cohortes_possibles = [annee for annee in annees_disponibles if annee + 3 in annees_disponibles]

plt.figure(figsize=(14, 10))

# Graphique principal : √©volution des cohortes
for region in niveaux_reset['region_academique'].unique():
    data_region = niveaux_reset[niveaux_reset['region_academique'] == region]

    for i, annee_debut in enumerate(cohortes_possibles):
        cohorte = suivre_cohorte(data_region, annee_debut)

        if len(cohorte) >= 2:  # Au moins 2 niveaux disponibles
            niveaux_ordre = ['6√®me', '5√®me', '4√®me', '3√®me']
            effectifs = [cohorte[niveau]['effectifs'] for niveau in niveaux_ordre if niveau in cohorte]
            positions = list(range(len(effectifs)))

            # Couleur diff√©rente par cohorte et style par r√©gion
            couleur = plt.cm.Set3(i / len(cohortes_possibles))
            style = '-' if region == 'Mayotte' else '--'

            plt.plot(positions, effectifs,
                    color=couleur, linestyle=style,
                    marker='o', linewidth=3, markersize=8,
                    label=f'{region} - Cohorte {annee_debut}',
                    alpha=0.8)

plt.title('√âvolution des cohortes d\'√©l√®ves par niveau\n(Suivi longitudinal)',
          fontsize=16, fontweight='bold')
plt.xlabel('Niveau scolaire', fontsize=12)
plt.ylabel('Nombre d\'√©l√®ves', fontsize=12)
plt.xticks(range(4), ['6√®me', '5√®me', '4√®me', '3√®me'])
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```
