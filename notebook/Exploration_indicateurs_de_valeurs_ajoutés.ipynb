### Heatmap de performance avec normalisation

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.impute import SimpleImputer
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Analyser les valeurs manquantes d'abord
def analyser_na(df, colonnes):
    """Analyse le pourcentage de NA par colonne"""
    na_analysis = {}
    for col in colonnes:
        na_count = df[col].isna().sum()
        na_percent = (na_count / len(df)) * 100
        na_analysis[col] = {
            'na_count': na_count,
            'na_percent': na_percent,
            'valeurs_manquantes': df[df[col].isna()].index.tolist()
        }
    return na_analysis

# S√©lectionner les indicateurs cl√©s
perf_cols = ['taux_de_reussite', 'taux_de_mentions', 'taux_d_acces_2nde_bac',
            'part_de_mentions_tb_avec_f√©licitations', 'part_de_mentions_tb_sans_f√©licitations',
            'part_de_mentions_b', 'part_de_mentions_ab']

df_perf_bac = df_region[perf_cols].copy()

# üîç ANALYSE DES VALEURS MANQUANTES
na_analysis = analyser_na(df_perf_bac, perf_cols)

print("üìä ANALYSE DES VALEURS MANQUANTES")
print("="*50)
for col, info in na_analysis.items():
    print(f"\n{col}:")
    print(f"  Valeurs manquantes: {info['na_count']}/{len(df_perf_bac)} ({info['na_percent']:.1f}%)")
    if info['valeurs_manquantes']:
        print(f"  R√©gions concern√©es: {', '.join(info['valeurs_manquantes'])}")

# üéØ STRAT√âGIE D'IMPUTATION
def imputation_intelligente(df, seuil_na=30):
    """
    Impute les valeurs manquantes selon une strat√©gie adapt√©e
    """
    df_imputed = df.copy()
    colonnes_imputees = []
    colonnes_supprimees = []

    for col in df.columns:
        na_percent = (df[col].isna().sum() / len(df)) * 100

        if na_percent == 0:
            # Pas de NA, rien √† faire
            continue
        elif na_percent <= seuil_na:
            # Imputation avec SimpleImputer
            print(f"‚úÖ Imputation de '{col}' ({na_percent:.1f}% NA)")

            # Essayer diff√©rentes strat√©gies selon la distribution
            if df[col].std() / df[col].mean() < 0.2:  # Faible variabilit√© -> m√©diane
                strategy = 'median'
            else:  # Forte variabilit√© -> moyenne
                strategy = 'mean'

            imputer = SimpleImputer(strategy=strategy)
            df_imputed[col] = imputer.fit_transform(df[[col]]).flatten()
            colonnes_imputees.append(f"{col} (strat√©gie: {strategy})")

        else:
            # Trop de NA, on supprime ou on traite sp√©cialement
            print(f"‚ö†Ô∏è  '{col}' a {na_percent:.1f}% NA (> {seuil_na}%) - Suppression recommand√©e")
            colonnes_supprimees.append(col)

    return df_imputed, colonnes_imputees, colonnes_supprimees

# Appliquer l'imputation
df_imputed, cols_imputees, cols_supprimees = imputation_intelligente(df_perf_bac, seuil_na=30)

print(f"\nüîß R√âSULTAT DE L'IMPUTATION")
print("="*35)
print(f"Colonnes imput√©es: {len(cols_imputees)}")
for col in cols_imputees:
    print(f"  ‚Ä¢ {col}")

if cols_supprimees:
    print(f"\nColonnes √† supprimer (trop de NA): {len(cols_supprimees)}")
    for col in cols_supprimees:
        print(f"  ‚Ä¢ {col}")

    # Supprimer les colonnes avec trop de NA
    df_final_clean = df_imputed.drop(columns=cols_supprimees)
else:
    df_final_clean = df_imputed

print(f"\nDataframe final: {df_final_clean.shape}")
```

```python
# üöÄ VERSION AVANC√âE AVEC CHOIX DE STRAT√âGIES
def imputation_avancee(df, seuil_na=30):
    """
    Imputation avanc√©e avec choix automatique de la meilleure strat√©gie
    """
    from sklearn.impute import KNNImputer
    from sklearn.experimental import enable_iterative_imputer
    from sklearn.impute import IterativeImputer

    df_results = {}
    strategies_info = {}

    for col in df.columns:
        na_percent = (df[col].isna().sum() / len(df)) * 100

        if na_percent == 0:
            df_results[col] = df[col].copy()
            strategies_info[col] = "Pas de NA"
            continue

        if na_percent > seuil_na:
            print(f"‚ö†Ô∏è {col}: {na_percent:.1f}% NA (ignor√©)")
            continue

        # Tester diff√©rentes strat√©gies
        strategies = {}

        # 1. SimpleImputer avec diff√©rentes m√©thodes
        for method in ['mean', 'median']:
            imputer = SimpleImputer(strategy=method)
            strategies[f'Simple_{method}'] = imputer.fit_transform(df[[col]]).flatten()

        # 2. KNN Imputer (si pas trop de colonnes)
        if df.shape[1] <= 10:
            try:
                knn_imputer = KNNImputer(n_neighbors=3)
                # Utiliser toutes les colonnes num√©riques pour KNN
                df_numeric = df.select_dtypes(include=[np.number])
                strategies['KNN'] = knn_imputer.fit_transform(df_numeric)[:, df_numeric.columns.get_loc(col)]
            except:
                pass

        # Choisir la meilleure strat√©gie (celle qui pr√©serve le mieux la distribution)
        original_std = df[col].std()
        best_strategy = None
        best_score = float('inf')

        for strategy_name, imputed_values in strategies.items():
            # Score bas√© sur la pr√©servation de l'√©cart-type
            new_std = pd.Series(imputed_values).std()
            score = abs(original_std - new_std)

            if score < best_score:
                best_score = score
                best_strategy = strategy_name
                df_results[col] = imputed_values

        strategies_info[col] = f"{best_strategy} (NA: {na_percent:.1f}%)"
        print(f"‚úÖ {col}: {best_strategy} choisie ({na_percent:.1f}% NA)")

    return pd.DataFrame(df_results, index=df.index), strategies_info

# Appliquer l'imputation avanc√©e
df_advanced_imputed, strategies_used = imputation_avancee(df_perf_bac)

print(f"\nüéØ STRAT√âGIES UTILIS√âES")
print("="*30)
for col, strategy in strategies_used.items():
    print(f"{col:<40} : {strategy}")
```

```python
# üé® HEATMAP AVEC DONN√âES IMPUT√âES
# Utiliser df_final_clean (ou df_advanced_imputed pour la version avanc√©e)
df_clean = df_final_clean  # ou df_advanced_imputed

# Normalisation
scaler = MinMaxScaler()
df_scaled_clean = pd.DataFrame(scaler.fit_transform(df_clean),
                              index=df_clean.index,
                              columns=df_clean.columns)

# Score global
df_scaled_clean['Score_global'] = df_scaled_clean.mean(axis=1)
df_final_sorted = df_scaled_clean.sort_values('Score_global', ascending=False)
df_heatmap_final = df_final_sorted.drop('Score_global', axis=1)

# Renommer les colonnes
colonnes_bac_clean = {
    'taux_de_reussite': 'Taux r√©ussite',
    'taux_de_mentions': 'Taux mentions',
    'taux_d_acces_2nde_bac': 'Acc√®s 2nde‚ÜíBac',
    'part_de_mentions_tb_avec_f√©licitations': 'TB f√©licitations',
    'part_de_mentions_tb_sans_f√©licitations': 'TB',
    'part_de_mentions_b': 'Bien',
    'part_de_mentions_ab': 'Assez Bien'
}

# Ne renommer que les colonnes qui existent
colonnes_a_renommer = {k: v for k, v in colonnes_bac_clean.items() if k in df_heatmap_final.columns}
df_heatmap_final = df_heatmap_final.rename(columns=colonnes_a_renommer)

plt.figure(figsize=(14, 12))

# Heatmap avec annotations des valeurs originales imput√©es
df_original_sorted = df_clean.reindex(df_final_sorted.index)
df_original_renamed = df_original_sorted.rename(columns=colonnes_a_renommer)

sns.heatmap(df_heatmap_final,
            cmap="RdYlGn",
            linewidths=0.8,
            annot=df_original_renamed.round(1).values,  # Valeurs originales (imput√©es)
            fmt='.1f',
            cbar_kws={'label': 'Score normalis√© (0-1)'},
            square=False,
            annot_kws={'size': 9})

plt.title("Performance au baccalaur√©at par r√©gion\n(Donn√©es imput√©es scientifiquement - Classement par score global)",
          fontsize=16, fontweight='bold', pad=20)
plt.xlabel("Indicateurs de performance", fontsize=14)
plt.ylabel("R√©gions acad√©miques", fontsize=14)
plt.xticks(rotation=45, ha='right', fontsize=12)
plt.yticks(rotation=0, fontsize=11)

# Ajouter une note sur l'imputation
plt.figtext(0.02, 0.02,
           f"Note: Imputation r√©alis√©e pour les colonnes avec <30% de valeurs manquantes\n" +
           f"Strat√©gies utilis√©es: {', '.join(set([s.split('(')[0].strip() for s in strategies_used.values() if 'NA:' in s]))}",
           fontsize=9, style='italic',
           bbox=dict(boxstyle='round', facecolor='lightyellow', alpha=0.8))

plt.tight_layout()
plt.subplots_adjust(bottom=0.15)  # Laisser place pour la note
plt.show()

# Nouveau classement
print("\nüèÜ CLASSEMENT AVEC IMPUTATION INTELLIGENTE")
print("="*45)
for i, (region, score) in enumerate(df_final_sorted['Score_global'].items(), 1):
    emoji = "ü•á" if i == 1 else "ü•à" if i == 2 else "ü•â" if i == 3 else f"{i:2d}."
    print(f"{emoji} {region:<30} {score:.3f}")
```

```python
from sklearn.preprocessing import MinMaxScaler, RobustScaler, StandardScaler

# Analysons d'abord les distributions de tes donn√©es
def analyser_distributions(df):
    """Analyse les distributions pour choisir le bon scaler"""
    stats = {}

    for col in df.columns:
        data = df[col].dropna()

        # Statistiques descriptives
        Q1 = data.quantile(0.25)
        Q3 = data.quantile(0.75)
        IQR = Q3 - Q1

        # D√©tection des outliers (m√©thode IQR)
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        outliers = data[(data < lower_bound) | (data > upper_bound)]

        # Coefficient de variation
        cv = data.std() / data.mean() if data.mean() != 0 else 0

        # Test de normalit√© (Shapiro-Wilk pour petits √©chantillons)
        from scipy import stats as scipy_stats
        if len(data) <= 50:
            _, p_value = scipy_stats.shapiro(data)
            is_normal = p_value > 0.05
        else:
            is_normal = None

        stats[col] = {
            'mean': data.mean(),
            'median': data.median(),
            'std': data.std(),
            'min': data.min(),
            'max': data.max(),
            'Q1': Q1,
            'Q3': Q3,
            'IQR': IQR,
            'outliers_count': len(outliers),
            'outliers_percent': (len(outliers) / len(data)) * 100,
            'cv': cv,
            'skewness': data.skew(),
            'is_normal': is_normal,
            'outliers_values': outliers.tolist()
        }

    return stats

# Analyser les donn√©es du bac
stats_bac = analyser_distributions(df_perf_bac.dropna())

print("üìä ANALYSE DES DISTRIBUTIONS")
print("="*50)
for col, stat in stats_bac.items():
    print(f"\n{col}:")
    print(f"  Outliers: {stat['outliers_count']} ({stat['outliers_percent']:.1f}%)")
    print(f"  Coefficient variation: {stat['cv']:.3f}")
    print(f"  Skewness: {stat['skewness']:.3f}")
    if stat['outliers_count'] > 0:
        print(f"  Valeurs aberrantes: {[f'{x:.1f}' for x in stat['outliers_values'][:3]]}")
```

```python
# Comparaison des 3 scalers principaux
def comparer_scalers(df):
    """Compare l'effet des diff√©rents scalers"""

    scalers = {
        'MinMaxScaler': MinMaxScaler(),
        'RobustScaler': RobustScaler(),
        'StandardScaler': StandardScaler()
    }

    # Pr√©parer les donn√©es (sans NaN pour cette d√©mo)
    df_clean = df.dropna()

    fig, axes = plt.subplots(3, len(df_clean.columns), figsize=(20, 12))

    for i, (scaler_name, scaler) in enumerate(scalers.items()):
        # Normaliser les donn√©es
        df_scaled = pd.DataFrame(
            scaler.fit_transform(df_clean),
            index=df_clean.index,
            columns=df_clean.columns
        )

        for j, col in enumerate(df_clean.columns):
            ax = axes[i, j]

            # Histogramme
            ax.hist(df_scaled[col], bins=10, alpha=0.7, color=f'C{i}')
            ax.set_title(f'{scaler_name}\n{col}', fontsize=10)
            ax.set_ylabel('Fr√©quence')

            # Statistiques
            ax.text(0.02, 0.98,
                   f'Min: {df_scaled[col].min():.2f}\n' +
                   f'Max: {df_scaled[col].max():.2f}\n' +
                   f'M√©diane: {df_scaled[col].median():.2f}',
                   transform=ax.transAxes,
                   verticalalignment='top',
                   bbox=dict(boxstyle='round', facecolor='white', alpha=0.8),
                   fontsize=8)

    plt.tight_layout()
    plt.suptitle('Comparaison des scalers sur les donn√©es du baccalaur√©at',
                 fontsize=16, fontweight='bold', y=1.02)
    plt.show()

# Appliquer la comparaison
comparer_scalers(df_perf_bac[['taux_de_reussite', 'taux_de_mentions', 'taux_d_acces_2nde_bac']])
```

### Analyse comparative Outre-mer vs M√©tropole

```python
# Identifier les r√©gions
outre_mer = ['GUADELOUPE', 'GUYANE', 'LA REUNION', 'MARTINIQUE', 'MAYOTTE']
metropole = [region for region in df_region.index if region not in outre_mer]

fig, axes = plt.subplots(2, 2, figsize=(16, 12))

# Indicateurs √† comparer
indicateurs = ['taux_de_reussite', 'taux_de_mentions', 'taux_d_acces_2nde_bac']
titres = ['Taux de r√©ussite (%)', 'Taux de mentions (%)', 'Acc√®s 2nde‚ÜíBac (%)']

for i, (indic, titre) in enumerate(zip(indicateurs, titres)):
    ax = axes[i//2, i%2]

    # Moyennes
    moy_metro = df_region.loc[metropole, indic].mean()
    moy_om = df_region.loc[outre_mer, indic].mean()

    # Boxplots
    data_metro = df_region.loc[metropole, indic]
    data_om = df_region.loc[outre_mer, indic]

    ax.boxplot([data_metro, data_om], labels=['M√©tropole', 'Outre-mer'],
               patch_artist=True,
               boxprops=dict(facecolor='lightblue', alpha=0.7),
               medianprops=dict(color='red', linewidth=2))

    ax.set_title(titre, fontsize=14, fontweight='bold')
    ax.grid(axis='y', alpha=0.3)

    # Ajouter les moyennes
    ax.text(1, moy_metro, f'Moy: {moy_metro:.1f}', ha='center',
           bbox=dict(boxstyle='round', facecolor='white', alpha=0.8))
    ax.text(2, moy_om, f'Moy: {moy_om:.1f}', ha='center',
           bbox=dict(boxstyle='round', facecolor='white', alpha=0.8))

# Graphique 4 : D√©tail par territoire OM
ax4 = axes[1, 1]
data_om_detail = df_region.loc[outre_mer, 'taux_de_reussite'].sort_values()
bars = ax4.barh(range(len(data_om_detail)), data_om_detail.values,
                color='#e74c3c', alpha=0.8)
ax4.set_yticks(range(len(data_om_detail)))
ax4.set_yticklabels(data_om_detail.index)
ax4.set_title('Taux de r√©ussite en Outre-mer', fontweight='bold')
ax4.grid(axis='x', alpha=0.3)

moyenne_metro = df_region.loc[metropole, 'taux_de_reussite'].mean()
ax4.axvline(x=moyenne_metro, color='blue', linestyle='--',
           label=f'Moyenne m√©tropole: {moyenne_metro:.1f}%')
ax4.legend()

plt.suptitle('Comparaison M√©tropole vs Outre-mer - Baccalaur√©at',
             fontsize=16, fontweight='bold')
plt.tight_layout()
plt.show()
```
